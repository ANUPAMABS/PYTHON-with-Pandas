```python
class data_analysis():
    
    def basic(self):
        df=pd.read_excel('Data Analysis Case Study Jun_2021.xlsx')
        print(df.shape)
        print(df.info())
        print(df.dtypes)
        print(df.isnull().sum())
    
    def count_columns(self):
        col=['Branch','City','Customer type', 'Gender', 'Product line', 'Day','Payment']
        
        for i in col:
            print(i," distinct count \n ", df[i].value_counts())
    
    def total_sale(self):
        t=round(df['Total'].sum(),2)
        print("Total sale amount ",t)
        print(df.groupby('Gender').Total.sum())
        
        df1=df[df.Gender=='Male'].Total.sum()
        mp = df1/t *100
        print("Total sale of male in percentage",mp)
        
        df2=df[df.Gender=='Female'].Total.sum()
        fp = df2/t *100
        print("Total sale of Female in percentage",fp)
        
    def branch_city(self):
        df3=df[["Branch","City"]]
        print(df3)
        print(df3.groupby('City').first())
        
    def city_sales(self):
        print("Citywise Sales amount",df.groupby('City').Total.sum())
        print("Maximum sale of the city ",df.groupby('City').Total.sum().max())
        print("Minimum sale of the city ",df.groupby('City').Total.sum().min())
    
    def payment(self):
        print("Payment method by male and Female ",df.groupby('Payment').size())
        print("Maximum payment",df.groupby('Payment').size().max())
        
        df4= df[['Gender','Payment']]
        print(df4)
        df5 = df4[df4.Gender=='Female']
        df6= df4[df4.Gender=='Male']
        print("Female  maximum Payment method is ",df5['Payment'].max())
        print("Male maximum Payment method is ",df6['Payment'].max())
    
    def week_sale(self):
        print("Total Sale amount for each day ",df.groupby('Day').Total.sum())
        print("Maximum sale amount of the week",df.groupby('Day').Total.sum().max())
        print("Total sale of each day",df.groupby('Day').size())
        
    def tax(self):
        print("Tax collected by government",df['Tax 5%'].sum())
        tc=df[['City','Tax 5%']]
        print("Citywise Tax collected ",tc.groupby('City').sum())
        print("Highest Tax collected ",tc.groupby('City').sum().max())
        print("Lowest Tax collected ",tc.groupby('City').sum().min())
        
    def analysis(self):
        df7=pd.pivot_table(df,index=['City'])
        print("PIVOT TABLE",df7)
        print("CORRELATION",df.corr())
        print("DATATYPES",df.dtypes)
        newdf=df.convert_dtypes()
        print("CONVERTED DATATYPES",newdf.dtypes)
        import matplotlib
        pc=df.groupby('City').Total.sum()
        print("PLOTING GRAPH OF CITY VS TOTAL SALE")
        pc.plot()
        d=df.groupby('Day').Total.sum()
        print("PLOTING GRAPH OF DAY VS TOTAL SALE")
        d.plot()
```

```python
fc=data_analysis()
fc.basic()
```
